#  What is Python? 
----
- Python is a popular, high-level programming language known for its simplicity and readability.
- It is used in many areas, including web development, data science, artificial intelligence, machine learning, scripting, automation, and more.
- Python's syntax is clear and easy to learn for beginners.

# Applications include:
 - Data analysis and visualization
 - Web and internet development
 - Machine learning and AI
 - Robotics and automation
 - Game development
 - Scripting and process automation

# Installing Python
- Visit the official Python website (https://python.org) and download the latest version for your operating system.
- After installation, verify in your command prompt or terminal:
- python --version
 or
 python3 --version

# Installing IDEs
- IDE stands for Integrated Development Environment, which helps you write, debug, and run code easily.
- Popular IDEs for Python:
### 1. Jupyter Notebook
    - Specially useful for data science, interactive code, explanations, and visualizations.
    - Install with: pip install notebook
    - Run with: jupyter notebook
### 2. Visual Studio Code (VS Code)
    - Lightweight, extensible editor.
    - Download from: https://code.visualstudio.com/
    - Install Python extension in VS Code for Python coding.

---- 
# Running Python Scripts 
- You can write Python scripts in files with the '.py' extension.
- To run a script, open a terminal and use:
 python script_name.py
 or
 python3 script_name.py
- You can also run code interactively in IDEs like Jupyter or the Python shell.

# Writing Your First Python Program


```python
# Let's write a simple program that prints a message to the screen.
print("Hello, World!")  # This will print 'Hello, World!' to the output.
```

# Python Interactive Shell Basics
- The interactive shell (also called REPL: Read-Eval-Print Loop) lets you execute Python commands one at a time.
- To start the shell, type `python` or `python3` in your command prompt or terminal.
### Example:
- > print("Hello from shell!")
- You will see: Hello from shell!
- You can use the shell to quickly test ideas, calculations, or debug code.
