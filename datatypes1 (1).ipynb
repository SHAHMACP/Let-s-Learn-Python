* STRING SLICING

```python
name="Python Programming"

print(name[0:9])                          # STARTING POSITION 0 AND STOP POSITION 9(EXCLUSIVE)
print(name[:])                             # IF STARTING AND STOP INDEX IS EMPTY THAT MEANS RETURN THE ENTIRE STRING

print(name[9])                            # RETURNS THE 9 TH CHARACTER
print(name[2])

print(name[3:])                            # STARTING FROM 3 POSITION TILL END
print(name[:7])                            # STARTING IS EMPTY SO IT WILL START FROM BEGINNING AND STOP AT 7 POSITION(7 position value will not be included)

print(name[0:-1])                          # START AT 0 TH POSITION AND STOP AT LAST INDEX(EXCLUSIVE)
print(name[-4:-1])

print(name[0:9:2])                         # START 0 STOP 9 STEP 2-IT WILL SKIP ONE ELEMENT
print(name[3:8:3])

print(name[::-1])
print(name[-3:-7:-1])
print(name[-2:-10:-2])
```

* STRING METHODS - Search and Check Methods

```python
text="hello welcome to python programming python"

print(text.startswith("hello"))                      # Checks if string starts with substring hello, return true or false

print(text.endswith("hello"))                       #   Checks if string ends with substring hello, return true or false

print(text.find("o"))                          # Returns index of first occurrence of substring (-1 if not found)
print(text.find("Z"))  
print(text.find("python"))

print(text.index('o'))                             # Returns index of first occurrence (raises error if not found)

print(text.count('o'))                            # Counts occurrences of substring 'o'
 
print(text.isalpha())                            # Checks if all characters are letters , if yes returns true otherwise false
print(text.isdigit())                             # Checks if all characters are digits, if yes returns true otherwise false
print(text.isspace())                             # Checks if all characters are whitespace, if yes returns true otherwise false

print(text.islower())                              # Checks if all characters are lowercase
print(text.isupper())                              # Checks if all characters are uppercase
print(text.isalnum())                               # Checks if all characters are alphanumeric
```

* STRING METHODS - Modification Methods

```python
text="    python  programming     "

print(text.strip())                        # REMOVES WHITESPACE FROM BOTH SIDES

print(text.rstrip())                       # REMOVES WHITESPACES FROM THE RIGHT SIDE

print(text.lstrip())                       # REMOVES WHITESPACES FROM THE LEFT SIDE

print(text.split(" "))                      # SPLIT A TEXT USING WHITESPACE
print(text.split("y"))                       # SPLITTING A TEXT USING THE SUBSTRING 'y'
print(text.split("p")) 

print(text.replace('p',' '))                # REPLACE EACH OCCURENCE OF substring 'p'  with whitespace
print(text.replace('python','good'))


text = "python"
result = "*".join(text)                           # Inserts '-' between each character
print(result)
```

7. LIST

* LIST IS AN ORDERED COLLECTION OF ELEMENTS

* IT CAN CONTAIN DIFFERENT TYPES OF VALUES LIKE INTEGERS, FLOATING POINT NUMBERS, BOOLEAN VALUES, STRING, NULL VALUES , COMPLEX VALUES.

* A LIST CAN CONTAIN ANOTHER LISTS

* LIST CAN CONTAIN DUPLICATE VALUES

* LIST IS CHANGEABLE (VALUES CAN BE CHANGED AFTER CREATION)

```python
l=[1,2,3,4,5,6]                   # list of numbers , only contain integer values
l1=[3,"hello",4.6,None,'A',3,True, False,4,"hi",[1,2,3,4],True,["hi","hoi"],2+1j]         # LIST CONTAINING INTEGERS , FLOATING POINT VALUES,NULL, BOOLEAN VALUES,LISTS,STRINGS
print(l1)
print(type(l1))
print(l)
print(type(l))
```

* LIST SLICING

* Same as String slicing

```python
l=["hello","hi",23,45.6,True,12,78,False,[1,2,3,4,5,6]]

print(l[3])                                    # IT WILL RETRIEVE THE VALUE FROM THE POSITION 6
print(l[1])                                    # IT WILL RETRIEVE THE VALUE FROM THE POSITION 1

print(l[-1])                                   # IT WILL RETRIEVE THE VALUE FROM THE LAST POSITION (NEGATIVE INDEXING)
print(l[-2])                                   # IT WILL RETRIEVE THE VALUE FROM THE POSITION -2

print(l[2:4])                                  # IT STARTS FROM 2 POSITION AND END AT 3  (OUTPUT: [23,45.5], 4 th position value will not be included)
print(l[-5:-1])

print(l[-1:-6:-1])
print(l[::-1])                                 # IT INCLUDES THE ENTIRE STRING (BACKWARD)
print(l[-1::-1])                               # IT INCLUDES THE ENTIRE STRING  (BACKWARD)
print(l[:-1])

print(l[-1][0:3])                      # Accessing the elements of inner list
```

* CHANGING LIST ELEMENTS

                    * It can be done using item assignment

```python
ls=["hello","hi",23,45.6,True,12,78,False]

ls[0]=34                    # 0th value will be changed from "hello" to 34
print(ls)

ls[-1]="python programming"  # Last item will be changed to the given string
print(ls)
```

* LIST METHODS

* To add new elements into the list : append(element) , insert(position,element) , extend()

                * append() : To insert a single element , the element will be inserted at the last position 

```python
lis=[23,45,78,90,"hello",True]

lis.append(89)
print(lis)

lis.append("hi")
print(lis)

lis.append([1,2,3])
print(lis)
```

               * insert()  - inserting an element into the specified position

```python
lis=[23,45,78,90,"hello",True]

lis.insert(0,"hello")                         # inserting "hello" to the 0 th position
print(lis)

lis.insert(-1,False)
print(lis)

lis.insert(-2,[1,2,3,4])
print(lis)
```

              * extend() - To extend a list with the elements of another list or tuple

```python
l1=[1,2,3,4,"hello"]
l2=[5,6,7,8,9]
t=(10,11,12,13)

l1.extend(l2)             # added the elements of l2 to the list l1
print(l1)

l1.extend(t)              # added the elements of the tuple t to the list l1
print(l1)

```

* LIST METHODS- To remove the elements from the list

         *  remove(element)  : To remove the specified element

```python
l1=[1,2,3,4,"hello",34,2,2,1]

l1.remove("hello")                        #removes "hello" from the list
print(l1)

l1.remove(1)                             # list contains two 1 s, and remove will removes the first occurence of 1
print(l1)


```

        * pop(position)- deletes the value from the specified position

```python
l1=[1,2,3,4,"hello",34,2,2,1]

l1.pop(1)                             # removes the element from the first position
print(l1)

l1.pop(-1)                            # removes the element from the last position
print(l1)

l1.pop()                              # if we do not specify the position , then it will remove the value from the last position
print(l1)
```

            * del keyword   - It can be used to delete a specified item and also to delete an entire list 

```python
l1=[1,2,3,4,"hello",34,2,2,1]

del l1[0]                            # removes the element from the 0 th position
print(l1)


del l1[3]                            # removes the element from the 3 rd position 
print(l1) 
```

```python
del l1                          # delete the entire list
```

            * clear() - To delete only the values of the list, list will be there. 

```python
l1=[1,2,3,4,"hello",34,2,2,1]

l1.clear()
print(l1)
```

* Sorting a List - sort() method

```python
l1=[34,2,67,12,90,44]
l1.sort()                                # SORTING THE LIST IN ASCENDING ORDER
print(l1)

l2=["rose","jasmine","hibiscus","lotus","orchid","marigold"]
l2.sort()          # sorting strings in ascending order (A-Z)
print(l2)

l3=["rose","Jasmine","hibiscus","lotus","Orchid","marigold"]
l3.sort()           # By default the sort() method is case sensitive, 
print(l3)           # resulting in all capital letters being sorted before lower case letters
```

```python
l1=[34,2,67,12,90,44]
l1.sort(reverse=True)                                # SORTING THE LIST IN DESCENDING ORDER
print(l1)

l2=["rose","jasmine","hibiscus","lotus","orchid","marigold"]
l2.sort(reverse=True)          # sorting strings in descending order (Z-A)
print(l2)

l3=["rose","Jasmine","hibiscus","lotus","Orchid","marigold"]
l3.sort(reverse=True)            
print(l3)           
```

* Copying a List - copy() method

```python
l1=[10,20,30,40,50]
l2=l1                 # copying, but changes made in original list will affect the copied list
print("Original list:",l1)
print("Copied list:",l2)

l1[0]=16
print("Original list:",l1)
print("Copied list:",l2)
print("Changes made in list1 affected the list2. Both changed")

```

```python
l1=[10,20,30,40,50]
l2=l1.copy()
print("Original list:",l1)
print("Copied list:",l2)


l1[0]=16
print("Original list:",l1)
print("Copied list:",l2,"\n")
print("Changes made in list1 not affected the list2.")
```

8. TUPLE

* Tuples are used to store multiple items in a single variable.

* A tuple is a collection which is ordered and unchangeable.

* Tuples are written with round brackets ( ).

* Python tuple is immutable.

* allow duplicate values.

* Creating a tuple:

         * Tuple items can be of any data type

         
         * To create a tuple with only one item, you have to add a comma after the item,   otherwise Python willnot recognize it as a tuple.

```python
t=(1,2,3,"hello",[1,2],(1,5),"hi",1,True) 
print(type(t))
print(t)
```

* Dont create a tuple of one element without adding a comma after element, because it will not be considered as tuple by python .

```python
t=(1)                       # not a tuple, its an int
print(type(t))
tnew=(1,)
print(type(tnew))


t1=("hello")                 # not a tuple , its string
print(type(t1))
t1new=("hello",)
print(type(t1new))


t2=(True)                    # not a tuple, its bool
print(type(t2))
t2new=(True,)
print(type(t2new))


t3=([1,2,3])
print(type(t3))
t3new=([1,2,3],)
print(type(t3new))
```

* It is also possible to use the tuple() constructor to make a tuple.

```python
l=[1,2,3,4,5]
t=tuple(l)              # list will be converted to tuple
print(t)

t1=tuple((1,2,3,4,5,6,7,8,9,"hello"))
print(t1)

t2=tuple({1,4,6,8,"hi"})
print(t2)

t3=tuple(["hello","hi",2+8j,False,1])
print(t3)
```
